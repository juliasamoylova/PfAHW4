## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

Построим heatmap для визуализации наличия и отсутствия линейной зависимости параметров друг от друга

```python
plt.figure(figsize=(7, 7))
upp_mat = np.triu(df.corr())
sns.heatmap(corr_matrix, annot=True, linewidths=.5,
            cmap='coolwarm', mask= upp_mat);
```

Количество этажей и количество жилой порщади находятся в самой сильной линейной зависимости - 0.9
Также в линейной зависимости:
* 0.8 - жилая площадь и количество ванных комнат
* 0.8 грейд сильнее всего зависит от жилой площади, площади над землей и жилой площадью у соседей
* 0.7 - цена больше всего зависит от жилой площади и грейда
* 0.4-0.7 грейд также зависит от количества этажей, года постройки, количества ванных
* цена также зависит от соседей, количества ванных и вида
* есть небольшая зависимость грейда от года постройки
* общая площадь коррелирует с общей площаью соседних домов - дома примерно одного уровня находятся рядом


```python
data = df.groupby('yr_built')['price'].mean().reset_index()
data
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(data['yr_built'], data['price'])
plt.title('Распределение средних цен в зависимости от года постройки');
```

Вывод:

- распределение цены в зависимости от года постройки имеет U-образную форму;
- цена на дома, построенные с 1900-х до 1960-х, постепенно опускалась;
- при этом цена на дома, построенные в1940-х годах, принимает неожиданно низкие значения;
- цена на дома, построенные с 1960-х годов, вновь начала расти;
- для домов, построенных в2020-х годах, цена почти такая же, как и для домов, построенных в 1900-х годах;
- могу предположить, что старые дома, которые стоят больше 1 века, имеют особую историческую значимость и поэтому их цена находится на одном уровне с ценами домов, построенных в 2000-х годах.

```python
plt.figure(figsize=(14, 10))

sns.boxplot(x=df['grade'], y=df['price'], whis=1.0)

plt.xlabel('Класс')
plt.ylabel('Цена')
plt.title('Распределение цен в зависимости от класса дома');
```

Зависимость цены от грейда становится особо заметной начиная с 7 класса и до 13. Между 12 и 13 грейдом самая большая разница в цене.

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5, hue=df['grade'])

plt.title('Распределение цен по береговой линии')
plt.xlabel('Цена')
plt.ylabel('Наличия набережной');
```

* Дома на набережной с грейдом 13 - не продаются
* Если дом на набережной, то его грейд минимум 4
* Разбег цен на дома с набережной меньше, чем без
* Дом с самым высоким грейдом 13 без набережной дешевле дома с грейдом 12, но с набережной

```python
plt.figure(figsize = (14,10))
sns.boxplot(x = df['price'], y = df['floors'].astype('str'), whis = 1.5)
plt.xlabel('Цена')
plt.ylabel('Этажность')
plt.title('Стоимость/Этажность дома');
```

Состояние дома больше влияет на цену. В среднем дома с оценкой 5 дороже, но некоторые дома с оценкой 4 и 3 могут быть даже дороже, зависит от других факторов, например - наличия набережной
